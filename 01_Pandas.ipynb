Import Finance Data (Historical prices) 

```python
# yfinance is used fetch data from Yahoo Finance 
import yfinance as yf
yf.pdr_override()
```

```python
# Inputs
symbol = "NFLX"
start = "2015-01-01"
end = "2023-02-22"

# Read data 
dataset = yf.download(symbol, start, end)

# View first 5 columns 
# DataFrame
dataset.head()
```

```python
# View the last 5 rows
dataset.tail()
```

```python
# Find mean of the first 4 rows
dataset.rolling(window=5).mean()
```

```python
# Find the standard deviation for the last 4 rows
dataset.rolling(window=5).std()
```

```python
# Isolate the standard deviations of the last 5 rows
(dataset.rolling(window=5).std()).tail()
```

```python
# Exponential weighted functions - mean (abbreviate it as ewm) - Weighted Moving Average
dataset["ewm"] = dataset["Adj Close"].ewm(span=20, min_periods=0, adjust=False,ignore_na=False).mean()
dataset.head()
```

```python
# iloc[] exmaples
dataset.iloc[0][0]
```

```python
# iat[] examples
dataset.iat[0,0]
```

```python
# Reset index 
df = dataset.reset_index()
df.head()
```

```python
# loc[] example
df.loc[0]["High"]
```

```python
df.iloc[0]
```

```python
# Specify column using iloc[]
df.iloc[0]["Volume"]
```

```python
# Specify example using at[]
df.at[0, "Volume"]
```

```python
# Remove column example
# Remove Date
new_df = df.drop(["Date"], axis=1)
new_df.head()
```

```python
# Add column example 
# Add Daily Return
df["DailyReturn"] = (new_df["Adj Close"]/new_df["Adj Close"].shift(1)).dropna()
new_column = df["DailyReturn"]
new_df["DailyReturn"] = new_column
new_df.head()
```

```python
# Move column example
# Move Date in the first column 
cols = new_df.columns.to_list()
cols.insert(0, cols.pop(cols.index("Date")))
cols
```

```python
# iloc[] function enables one to select a particular row in a dataset
new_df = new_df.iloc[:]
new_df.head()
```

```python
# Select row 3 of the above dataset
new_df.iloc[2]
```

```python
# Delete a Column 
# Delete Close 
del new_df["Close"]
```

```python
new_df.head()
```

```python
# Rename column 
# Rename Adj Close as Close 
new_df = new_df.rename(index=str, columns={"Adj Close": "Close"})
new_df.head()
```

```python
# Change  column names to uppercase
new_df.columns = new_df.columns.str.upper()
new_df.head()
```
